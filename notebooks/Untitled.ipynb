```python
import numpy as np
import matplotlib.pyplot as plt

from src.portfolio.efficient_frontier import simulate_portfolios, max_sharpe_portfolio, min_volatility_portfolio

mean_returns = np.array([0.12, 0.08, 0.10])
cov_matrix = np.array([
    [0.04, 0.01, 0.02],
    [0.01, 0.09, 0.03],
    [0.02, 0.03, 0.06]
])

results, weights = simulate_portfolios(mean_returns, cov_matrix, n_portfolios=5000, risk_free_rate=0.02)

max_sharpe = max_sharpe_portfolio(results, weights)
min_vol = min_volatility_portfolio(results, weights)

plt.figure(figsize=(10, 6))
plt.scatter(results[1], results[0], c=results[2], marker='o')
plt.xlabel("Volatility (Risk)")
plt.ylabel("Return")
plt.title("Efficient Frontier (Random Portfolios)")

plt.scatter(max_sharpe["volatility"], max_sharpe["return"], marker="*", s=200, label="Max Sharpe")
plt.scatter(min_vol["volatility"], min_vol["return"], marker="*", s=200, label="Min Volatility")

plt.legend()
plt.colorbar(label="Sharpe Ratio")
plt.show()
```
